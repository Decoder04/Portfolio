### Dropping missing value rows to avoid matplotlib errors

```python
raw_data_count = recent_grads.shape[0]
raw_data_count
```

```python
recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = recent_grads.shape[0]
cleaned_data_count
```

### Generating various plots for exploratory analysis

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample size vs. Median')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample size vs. Unemployment_rate')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full_time vs. Median')

ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Share Women vs. Unemployment_rate')

ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')

ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

### Observations:

1) There is a slight negative correlation between median salary and share of women. As share increases, median salary decreases

2) There is no other significant correlation found
